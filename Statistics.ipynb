```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("stats/pilot-unet-v1.csv")
```

```python
loss = np.array(df["loss"])
mu = loss.mean()
sigma = loss.std()
print("{},{}".format(mu, sigma))
count, bins, ignored = plt.hist(loss, 100, density=True)
plt.plot(bins, 1/(sigma * np.sqrt(2 * np.pi)) * np.exp( - (bins - mu)**2 / (2 * sigma**2) ), linewidth=2, color='r')
plt.show()
```

```python
means = np.array([loss[np.random.randint(low=1, high=100, size=len(loss))].mean() for _ in range(10000)])
mu = means.mean()
sigma = means.std()

count, bins, ignored = plt.hist(means, 100, density=True)
plt.plot(bins, 1/(sigma * np.sqrt(2 * np.pi)) * np.exp( - (bins - mu)**2 / (2 * sigma**2) ), linewidth=2, color='r')
plt.show()
print(mu, sigma)
```
